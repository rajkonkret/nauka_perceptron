cały przykład z nauką

```python
import numpy as np
X = np.array([[23,4],[18,3],[8,2],[200,30],[8,150],[30,350],[5,100],[10,200]])

print(X)

y = np.array([1,1,1,1,0,0,0,0])
print(y)

w1 = -2 # wagi
w2 = 0.1
b = 6 # bias

# Współczynnik uczenia

learning_rate = 0.01

# Liczba epok
epochs = 10

errors = []

# Pętla treningowa
for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        # Obliczanie sumy ważonej
        weighted_sum = X[i][0] * w1 + X[i][1] * w2 + b

        # Przewidywanie (brak funkcji aktywacji, używamy progu 0)
        # odwzorowanie naszego if'a
        prediction = 1 if weighted_sum > 0 else 0

        # Obliczanie błędu
        error = y[i] - prediction
        total_error += abs(error)
        print("error=", error)

        # Aktualizacja wag i biasu tylko jeśli wystąpił błąd
        if error != 0:
            w1 += learning_rate * error * X[i][0]
            w2 += learning_rate * error * X[i][1]
            b += learning_rate * error

        errors.append(total_error)

        # Przerwanie, jeśli brak błędów
        if total_error == 0:
            break

# Wyświetlenie ostatecznych wag i biasu
print(f"Waga w1: {w1}")
print(f"Waga w2: {w2}")
print(f"Bias: {b}")
```

```python
for element in X:

    # pierwsze wejscie perceptrona
    # print(element[0], end="=")
    predicted_w1 = element[0] * w1
    # print(predicted_w1)

    # drugie wejscie perceptrona
    # print(element[1], end="=")
    predicted_w2 = element[1] * w2
    # print(predicted_w2)

    predict_sum = predicted_w1 + predicted_w2 + b
    print(predict_sum, end="=")
    if predict_sum > 0:
        print("Obrączka", end=" : ")
        index = np.where((X == element).all(axis=1))[0]
        print(y[index])
    else:
        print("Długopis", end=" : ")
        index = np.where((X == element).all(axis=1))[0]
        print(y[index])
```

```python
import matplotlib.pyplot as plt
# Wykres błędu w funkcji epok
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(len(errors)), errors, marker='o')
plt.xlabel('Epoki')
plt.ylabel('Liczba błędów')
plt.title('Wykres błędu w funkcji epok')

# Wykres granicy decyzyjnej
plt.subplot(1, 2, 2)
for i in range(len(X)):
    if y[i] == 1:
        plt.scatter(X[i][0], X[i][1], color='blue', marker='o', label='Obrączka' if i == 0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='red', marker='x', label='Długopis' if i == 4 else "")

# Rysowanie linii decyzyjnej
x_values = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_values = -(w1 / w2) * x_values - (b / w2)
plt.plot(x_values, y_values, color='green', label='Granica decyzyjna')

plt.xlabel('Cecha 1')
plt.ylabel('Cecha 2')
plt.title('Granica decyzyjna perceptronu')
plt.legend()
plt.tight_layout()
plt.show()
```
